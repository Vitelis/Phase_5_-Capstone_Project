```python
## PREPARATION FOR MODELLING

#Importing additional libraries

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt


```

```python
# Preparing the data for time series analysis

# Considering 'Close' as the target variable

target_variable = 'Close'

# Creating a copy of the dataframe for time series analysis

time_series_df = historical_prices_df[['Date', target_variable]].copy()

# Set 'Date' as the index

time_series_df.set_index('Date', inplace=True)


# Train-test split

train_size = int(len(time_series_df) * 0.8)  # 80% for training, 20% for testing
train, test = time_series_df[:train_size], time_series_df[train_size:]
```

```python
#Visualizing ACF to get best p (AR order) 

from statsmodels.graphics.tsaplots import plot_acf

plot_acf(train[target_variable], lags=20)
```

```python
#creating a differencing series in order to identify the optimal value for d(Differencing order)

differenced_series = train[target_variable].diff().dropna()

differenced_series
```

```python
#visualizing  the PACF to identify the optimal value for q (MA order)

from statsmodels.graphics.tsaplots import plot_pacf

plot_pacf(train[target_variable], lags=20)
```

   ARIMA  MODELLING 

```python
# Importing library for ARIMA
from statsmodels.tsa.arima.model import ARIMA
```

```python
#sorting in ascending order 

time_series_df = time_series_df.sort_index()
```

```python
#setting the frequency of the date index.

#time_series_df.index = pd.DatetimeIndex(time_series_df.index, freq='A')
time_series_df.index = pd.to_datetime(time_series_df.index, format='%Y').to_period('A-DEC')
```

```python
# Training ARIMA model
arima_model = ARIMA(train[target_variable], order=(5,1,2))  
arima_fit = arima_model.fit()
```

```python
# Predict on the test set
arima_predictions = arima_fit.predict(start=len(train), end=len(train)+len(test)-1, typ='levels')
```

```python
# Evaluate the ARIMA model performance
arima_rmse = sqrt(mean_squared_error(test[target_variable], arima_predictions))
print(f'ARIMA Root Mean Squared Error (RMSE): {arima_rmse}')
```

```python
# Visualizing ARIMA predictions against the actual values

plt.figure(figsize=(14, 5))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(arima_predictions, label='ARIMA Predictions', linestyle='dashed')
plt.title('Safaricom Stock Price Forecasting using ARIMA')
plt.xlabel('Time in  Years ')
plt.ylabel('Closing Price (Ksh)')
plt.legend()
plt.show()
```
